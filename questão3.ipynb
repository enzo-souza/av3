```python
import pandas as pd
import matplotlib.pyplot as plt

# Lê o arquivo CSV, usando ; como separador
df = pd.read_csv('ITU_GCI_GCI_OVRL_SCRE_WIDEF.csv', sep=';')

# Converte a coluna OBS_VALUE para número, removendo pontos e tratando como float
df['OBS_VALUE'] = df['OBS_VALUE'].str.replace('.', '', regex=False).astype(float)

# Plota a tendência
plt.figure(figsize=(50,10))
plt.plot(df['REF_AREA_LABEL'], df['OBS_VALUE'], marker='o')
plt.title('Tendência de OBS_VALUE por País')
plt.xlabel('País')
plt.ylabel('OBS_VALUE')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Lê o arquivo CSV do continente, usando ; como separador
df_continente = pd.read_csv('ITU_GCI_GCI_OVRL_SCRE_WIDEF_CONTINENTE.csv', sep=';')

# Se OBS_VALUE_MEDIA já é numérica, não precisa converter. 
# Caso contrário, remova apenas vírgulas (decimal) e troque por ponto, se necessário:
if df_continente['OBS_VALUE_MEDIA'].dtype == object:
    df_continente['OBS_VALUE_MEDIA'] = df_continente['OBS_VALUE_MEDIA'].str.replace(',', '.', regex=False)
    df_continente['OBS_VALUE_MEDIA'] = df_continente['OBS_VALUE_MEDIA'].astype(float)

# Plota a tendência por continente
plt.figure(figsize=(10,5))
plt.plot(df_continente['CONTINENTE'], df_continente['OBS_VALUE_MEDIA'], marker='o')
plt.title('Tendência de OBS_VALUE_MEDIA por Continente')
plt.xlabel('Continente')
plt.ylabel('OBS_VALUE_MEDIA')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
