x = 2 cos(t) + cos(2t)
y = 2 sen(t) - sen(2t)

```python
import math
import matplotlib.pyplot as plt

# Define o valor de pi usando a biblioteca math
pi = math.pi

# Cria uma lista vazia para armazenar os valores de 'a'
a_vals = []
x_vals = []
y_vals = []


# Inicializa a variável val com zero
val = 0
x = 0
# Loop para preencher a lista 'a_vals' com valores de 0 até pi, incrementando de 0.1 em 0.1
while val <= 2*pi:
    a_vals.append(val)
    val += 0.1
    x = 2*math.cos(val) + math.sin(2*val)
    y = 2*math.sin(val) + math.cos(2*val)

    x_vals.append(x)
    y_vals.append(y)




# Plota os valores da lista 'a_vals'
plt.plot(a_vals)

# Adiciona o título ao gráfico
plt.title('Plot da variável a')

# Define o rótulo do eixo x
plt.xlabel('Índice')

# Define o rótulo do eixo y
plt.ylabel('Valor de a')

# Exibe o gráfico na tela
plt.show()

# Plota os valores da lista 'x_vals'
plt.plot(x_vals)
plt.title('Plot da variável x')
plt.xlabel('Índice')
plt.ylabel('Valor de y')
plt.show()

# Plota os valores da lista 'x_vals'
plt.plot(y_vals)
plt.title('Plot da variável y')
plt.xlabel('Índice')
plt.ylabel('Valor de y')
plt.show()
```
